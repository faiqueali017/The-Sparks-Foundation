# Step3: Using Elbow graph to find optimum nr. of clusters

#### Finding WCSS(Within-Clusters-Sum-of-Squares)

```python
# Converting dataframe to numpy array
x_data = data.iloc[:,[1,2,3,4]].values
```

```python
x_data[:5]
```

```python
x_data.shape
```

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x_data)
    # Interia method returns wcss for that model
    wcss.append(kmeans.inertia_)
    
wcss
```

#### Plotting Graph

```python
plt.figure(figsize=(10,5))

sns.set(style='whitegrid')
sns.lineplot(range(1,11), wcss, marker='o', color='red')

plt.title(ELBOW_PLT_TITLE)
plt.xlabel(COL_NR_CLUSTERS)
plt.ylabel(COL_WCSS)

plt.show()
```

The optimum value for K would be `k=3`. As we can see that with an increase in the number of cluster the WCSS value decreases. We select the value for K on the basis of the rate of decrease in WCSS and we can see that after 3 the drop in wcss is minimal.


# Step 4: Model Training

```python
kmeans = KMeans(n_clusters=3)

# Output of clusters
y_kmeans = kmeans.fit_predict(x_data)
```

```python
y_kmeans
```

# Step 5: Visualizing the Cluster

```python
plt.figure(figsize=(10,7))

# Visualizing the cluster - On the first two columns
plt.scatter(x_data[y_kmeans == 0,2], x_data[y_kmeans == 0,3], s=100, c='seagreen', alpha=0.8, label='Iris-versicolor')
plt.scatter(x_data[y_kmeans == 1,2], x_data[y_kmeans == 1,3], s=100, c='purple',alpha=0.8, label='Iris-setosa')
plt.scatter(x_data[y_kmeans == 2,2], x_data[y_kmeans == 2,3], s=100, c='yellow',alpha=0.8, label='Iris-virginica')

# Plotting the centriods of the clusters
plt.scatter(kmeans.cluster_centers_[:,2], 
            kmeans.cluster_centers_[:,3], 
            s=300,linewidths=5,
            c='red', marker='+', 
            label='Centriods')

plt.title(KMEANS_PLT_TITLE, fontweight='bold', fontsize=20)
plt.xlabel(COL_PETAL_W)
plt.ylabel(COL_PETAL_L)

plt.legend()
```

# Step 6: Comparing & Evaluating our Model

```python
# Extracting 'Species' col from original dataset
feature_labels = data['Species']

# Now replacing features labels to the predicted values format 
feature_labels = feature_labels.replace({'Iris-setosa': 1, 'Iris-versicolor': 0,'Iris-virginica':2})

# Converting feature_labels DF to numpyarray
feature_labels = feature_labels.to_numpy()
```

```python
prediction = kmeans.predict(x_data)

# Calculating the accuracy of our model
accuracy = metrics.accuracy_score(prediction, feature_labels)
```

```python
print(f'The accuracy of our model is: {accuracy:0.3%}')
```
